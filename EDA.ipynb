## Data Cleaning and Removing unnecessary columns

```python
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = data[col].fillna('Unknown')

    else:
        data[col] = data[col].fillna(data[col].median())
```

```python
data = data.drop('id', axis=1)
```

```python
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
```

## EDA and Visualization

```python
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(data[col], kde=True, bins=20, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.show()

    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=data[col], palette='Set2')
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 6. Correlation Matrix (Numeric Only)
# ----------------------------
plt.figure(figsize=(10,8))
sns.heatmap(data[numeric_cols + ['Depression']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix (Numeric Features)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=data['Depression'], palette='Set1')
plt.title('Distribution of Depression (Target)')
plt.show()
```

## Summary

### Dataset Structure and Quality
The dataset looks clean and easy to work with. Most columns have complete data, and only a few small gaps appear here and there. The numeric values mostly stay within normal ranges, with only a couple of points standing out as outliers, which is expected in real data.

### Target Variable Distribution
The Depression column isnt perfectly balanced. One category appears more often than the other. Because of this, splitting the data carefully or using balancing methods might help later when building models.

### Numerical Feature Insights
The graphs show that many numeric columns dont follow a perfect normal shape. Some have skewness, and a few contain mild outliers. The correlation heatmap didnt show any very strong relationships between numeric features, so theres no multicollinearity issue. However, a few moderate correlations may still help the model.

### Categorical Feature Patterns
The countplots show that the number of responses for each category is not evenly spread, which is common in surveys. When comparing these features with the depression levels, some categories clearly differ between the two groups. This means certain lifestyle or academic factors might be related to depression and could help the model learn better.

### Relationship Between Features and Target

1. Numeric vs Depression: Some numerical features show different distributions when split by depression level. This suggests they might influence the prediction.

2. Categorical vs Depression: A few categorical features also show clear differences between depression categories, meaning they may carry useful signals.

### Missing Values Visualization
The heatmap shows that missing values are very few and scattered. Theres no major pattern of missingness, and no column appears to have serious data gaps.

### Overall Assessment
Overall, the dataset is fit for modeling. Both numerical and categorical features seem helpful, and no single variable dominates the dataset. The model will likely learn from a combination of several small patterns. The slight imbalance in the depression column should be handled during model training.
